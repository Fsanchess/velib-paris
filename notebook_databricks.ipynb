```
#Dans cette analyse, nous calculons la capacité moyenne, maximale et minimale pour chaque arrondissement, et nous les représentons dans un graphique à barres pour une visualisation plus facile. Cela pourrait aider à identifier les arrondissements où la capacité est régulièrement saturée ou insuffisante, et à décider où ajouter ou réaffecter des racks de stationnement.

import matplotlib.pyplot as plt
import pyspark.sql.functions as F

# Read in the data from a file or database table
data = spark.sql("SELECT * FROM velib_data_1_csv")

# Group the data by arrondissement
grouped_arrondissements = data.groupBy('nom_arrondissement_communes')

# Calculate the average capacity for each arrondissement
avg_capacity = grouped_arrondissements.agg(F.avg('capacity')).withColumnRenamed('avg(capacity)', 'avg_capacity')

# Calculate the maximum capacity for each arrondissement
max_capacity = grouped_arrondissements.agg(F.max('capacity')).withColumnRenamed('max(capacity)', 'max_capacity')

# Calculate the minimum capacity for each arrondissement
min_capacity = grouped_arrondissements.agg(F.min('capacity')).withColumnRenamed('min(capacity)', 'min_capacity')

# Join the data frames
capacity_data = avg_capacity.join(max_capacity, on='nom_arrondissement_communes').join(min_capacity, on='nom_arrondissement_communes')

# Convert the resulting Spark DataFrame to a Pandas DataFrame
capacity_data_pd = capacity_data.toPandas()

# Plot the results using a bar chart
capacity_data_pd.plot(kind='bar', x='nom_arrondissement_communes', y=['avg_capacity', 'max_capacity', 'min_capacity'], color='#FFCC99')
plt.xlabel('Arrondissement')
plt.ylabel('Capacité')
plt.title('Capacité par Arrondissement')
plt.show()
```

```
import matplotlib.pyplot as plt
import pyspark.sql.functions as F

# Read in the data from a file or database table
data = spark.sql("SELECT * FROM velib_data_1_csv")

# Group the data by arrondissement
grouped_arrondissements = data.groupBy('nom_arrondissement_communes')

# Calculate the average capacity for each arrondissement
avg_capacity = grouped_arrondissements.agg(F.avg('capacity')).withColumnRenamed('avg(capacity)', 'avg_capacity')

# Calculate the maximum capacity for each arrondissement
max_capacity = grouped_arrondissements.agg(F.max('capacity')).withColumnRenamed('max(capacity)', 'max_capacity')

# Calculate the minimum capacity for each arrondissement
min_capacity = grouped_arrondissements.agg(F.min('capacity')).withColumnRenamed('min(capacity)', 'min_capacity')

# Join the data frames
capacity_data = avg_capacity.join(max_capacity, on='nom_arrondissement_communes').join(min_capacity, on='nom_arrondissement_communes')

# Convert the resulting Spark DataFrame to a Pandas DataFrame
capacity_data_pd = capacity_data.toPandas()

# Plot the results using a stacked area chart
ax = capacity_data_pd.plot.area(x='nom_arrondissement_communes', y=['avg_capacity', 'max_capacity', 'min_capacity'], color=['#FFCC99', '#FF9966', '#FF6600'], stacked=True)
ax.set_xlabel('Arrondissement')
ax.set_ylabel('Capacité')
ax.set_title('Capacité par Arrondissement')
# Set x-axis tick labels to rotate vertically and align properly
plt.xticks(rotation=90, ha='center')
# Adjust spacing to show all labels
plt.subplots_adjust(bottom=0.2)
plt.show()

```
